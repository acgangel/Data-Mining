# Workflow
> 想法：個別將兩個原始資料都整理好成tidy data，再合併成單一的tidy data  

    1. 載入資料與套件，並檢閱資料是否有亂碼、空值等問題
    2. 利用melt將不同類別的資料分開提取與轉換
    3. 資料清整好利用merge合併
    4. 刪除資料表內重複欄位、字串，排序輸出成tidy data
    5. 比照上方流程處理第二份檔案
    6. 將兩份整理好的資料表合併成tidy data
    7. 清整資料、刪除重複欄位、並排序輸出
    8. 檢查是否還有可以整合的變數欄位
    9. 繼續使用melt拆分資料表成數個資料表，並分開處理
    10. 將各資料表清整好後，利用concat合併

### 處理食品餐飲類資料
#### 載入套件，先讀入食品餐飲類資料做預處理，檔案內不僅有ASCII，故要再設定編碼才能順利讀取檔案

```python
import pandas as pd
df = pd.read_csv(r'D:\DataMining\week01\food.CSV',encoding='Big5')
df.head(5) #顯示前五項
```

```python
df.info() #檢閱資料
```

#### 利用melt格式轉換；將類別之後的所有欄位依照「年齡對應筆數」與「年齡對應金額」隔行提取成兩個資料表再做合併

```python
freq_df = pd.melt(df, id_vars=['年月','類別'], value_vars=list(df.columns)[2::2],
             var_name='年齡x', value_name='筆數(食)') #隔行提取
amount_df = pd.melt(df, id_vars=['年月'], value_vars=list(df.columns)[3::2],
             var_name='年齡y', value_name='金額(食)') #隔行提取
food_df = pd.merge(freq_df, amount_df, on='年月') #合併
food_df.head(10)
```

#### 僅留下一個年齡欄位，刪除年齡欄位內重複字串，再依照年月及年齡做排序，最後匯出tidy data

```python
food_df = food_df.drop(['年齡y'], axis=1) #刪除年齡_y
food_df['年齡x'] = food_df['年齡x'].map(lambda x: x.rstrip('[筆數]')) #刪除重複字串
food_df = food_df.sort_values(ascending=True, by=["年月","年齡x"]) #排序
food_df = food_df.rename({'年齡x': '年齡'}, axis=1) #重新命名
food_df.to_csv(r'D:\DataMining\week01\food-tidy.CSV',encoding='utf_8_sig', index=False) #輸出並編碼
food_df.head(10)
```

### 讀入服飾類資料做前處理
#### _同上的處理步驟_

```python
df = pd.read_csv(r'D:\DataMining\week01\cloth.CSV',encoding='Big5')
df.head(5)
```

```python
df.info() #檢閱資料是否有空值等狀況
```

```python
freq_df = pd.melt(df, id_vars=['年月','類別'], value_vars=list(df.columns)[2::2],
             var_name='年齡x', value_name='筆數(衣)')
amount_df = pd.melt(df, id_vars=['年月'], value_vars=list(df.columns)[3::2],
             var_name='年齡y', value_name='金額(衣)')
cloth_df = pd.merge(freq_df, amount_df, on='年月')
cloth_df.head(10)
```

```python
cloth_df = cloth_df.drop(['年齡y'], axis=1)
cloth_df['年齡x'] = cloth_df['年齡x'].map(lambda x: x.rstrip('[筆數]'))
cloth_df = cloth_df.sort_values(ascending=True, by=["年月","年齡x"])
cloth_df = cloth_df.rename({'年齡x': '年齡'}, axis=1)
cloth_df.to_csv(r'D:\DataMining\week01\cloth-tidy.CSV',encoding='utf_8_sig', index=False)
cloth_df.head(10)
```

```python
import sys

total_df = pd.merge(food_df, cloth_df, on='年月')
total_df = total_df.drop(['年齡_y'], axis=1)
total_df = total_df.sort_values(ascending=True, by=["年月","年齡_x"])
total_df = total_df.rename({'年齡_x': '年齡'}, axis=1)

cols=['年月','年齡','類別_x','筆數(食)','金額(食)','類別_y','筆數(衣)','金額(衣)']
total_df=total_df.ix[:,cols]

total_df.to_csv(r'D:\DataMining\week01\total-tidy.CSV',encoding='utf_8_sig', index=False)
total_df.head(10)
```

### 類別欄位又能再度提取，依照「類別」、「筆數」、「金額」拆成三個表單
#### 先對「類別」資料表做處理

```python
df = total_df
catetidy_df = pd.melt(df, id_vars=['年月','年齡'], value_vars=['類別_x','類別_y'],
             var_name='類別', value_name='分類')
casetidy_df = pd.melt(df, id_vars=['年月','年齡'], value_vars=['筆數(食)','筆數(衣)'],
             var_name='筆數分類', value_name='筆數')
amounttidy_df = pd.melt(df, id_vars=['年月','年齡'], value_vars=['金額(食)','金額(衣)'],
             var_name='金額分類', value_name='金額')
catetidy_df.head(10)
```

```python
catetidy_df = catetidy_df.drop(['類別'], axis=1) #方便後續三表做橫向合併，將類別欄位刪除
catetidy_df.head(10)
```

#### 接著「筆數類別」資料表做處理

```python
casetidy_df.head(10)
```

```python
casetidy_df = casetidy_df.drop(['年月','年齡','筆數分類'], axis=1) #方便後續三表做橫向合併，將前三欄欄位都刪除
casetidy_df.head(10)
```

#### 最後對「金額類別」資料表做處理

```python
amounttidy_df.head(10)
```

```python
amounttidy_df = amounttidy_df.drop(['年月','年齡','金額分類'], axis=1) #方便後續三表做橫向合併，將前三欄欄位都刪除
amounttidy_df.head(10)
```

### 使用concat對照Index將三個資料表合併，輸出成最終的tidy data

```python
final_df = pd.concat([catetidy_df, casetidy_df, amounttidy_df], axis=1)
final_df
```

```python
final_df.to_csv(r'D:\DataMining\week01\final-tidy.csv',encoding='utf_8_sig', index=False)
final_df.to_csv(r'D:\DataMining\week01\final-tidy.txt',encoding='utf_8_sig', index=False)
#輸出的資料量太多，若excel無法完整開啟csv，請讀取txt檔
```
